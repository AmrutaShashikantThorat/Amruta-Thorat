```python
# line plot
plt.figure(figsize=(15,9))
sns.lineplot(x='Age',y='StudyTimeWeekly',data=df,marker='o',mfc='red',color='black',ms=12)
plt.title('Line Plot')
plt.xlabel('Age')
plt.ylabel('StudyTimeWeekly')

plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.countplot(data=df,x='Absences',edgecolor='black')
plt.title('Bar plot')
plt.xlabel('Absences')
plt.ylabel('StudyTimeWeekly')
plt.xticks(range(1,40,10))
plt.show()
```
