```python
# line plot
plt.figure(figsize=(15,9))
sns.lineplot(x='Application order',y='Marital status',data=df,marker='o',mfc='red',color='black',ms=12)
plt.title('Line Plot')
plt.xlabel('AgeApplication order')
plt.ylabel('Marital status')

plt.grid(True)
plt.show()
```

```python
# join plot
sns.jointplot(x='Age at enrollment',y='GDP',data =df,hue=df['Gender']);
```

```python
plt.figure(figsize=(6,6))
sns.countplot(data=df,x='Age at enrollment',edgecolor='black')
plt.title('Bar plot')
plt.xlabel('Age at enrollment')
plt.ylabel('StudyTimeWeekly')
plt.xticks(range(1,40,10))
plt.show()
```
